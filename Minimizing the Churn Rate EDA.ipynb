# Data visualisation

```python
data.columns.values
```

```python
data2 = data.drop(columns=['user','churn'])
```

```python
%%time
plt.figure(figsize=[18,15])
plt.suptitle('Histogram for numerical columns', fontsize = 14)

for i in range(1, data2.shape[1] + 1):
    plt.subplot(10, 3, i)
    f = plt.gca()
    f.set_title(data2.columns.values[i - 1])
    vals = np.size(data2.iloc[:, i - 1].unique())
    plt.hist(data2.iloc[:, i - 1], bins= vals, ec = 'black', color='#48dbfb')
    plt.yticks([])
    
plt.tight_layout(rect=(0.4, -0.4, 1, 0.95))
plt.show()
```

# Pie chart

```python
## Pie Plots
data2 = data[['housing', 'is_referred', 'app_downloaded',
                    'web_user', 'app_web_user', 'ios_user',
                    'android_user', 'registered_phones', 'payment_type',
                    'waiting_4_loan', 'cancelled_loan',
                    'received_loan', 'rejected_loan', 'zodiac_sign',
                    'left_for_two_month_plus', 'left_for_one_month', 'is_referred']]
```

```python
%%time
fig = plt.figure(figsize= [18,15])
plt.suptitle('Pie chart distribution', fontsize= 14)

for i in range(1, data2.shape[1] + 1):
    plt.subplot(6,3,i)
    f = plt.gca()
    f.axes.get_yaxis().set_visible(False)
    f.set_title(data2.columns.values[i - 1])
    values = data2.iloc[:, i - 1].value_counts(normalize = True).values
    index = data2.iloc[:, i-1].value_counts(normalize = True).index
    plt.pie(values, labels= index, autopct='%1.1f%%')
plt.tight_layout(rect=(0.4, -0.4, 1, 0.95))  
plt.show()
```

# Correlation with Response Variable

```python
data.drop(columns = ['housing', 'payment_type','churn','user',
                         'registered_phones', 'zodiac_sign']).corrwith(data.churn).plot.bar(
                figsize=[15,10], title= 'Correlation with Response Variable', grid =True,
                fontsize= 12, rot = 45, cmap = 'bwr')
plt.show()
```

# Correlation matrix

```python
corr = data.drop(columns=['user', 'churn']).corr()
corr
```

```python
mask = np.zeros_like(corr)
tringle_indices = np.triu_indices_from(mask)
mask[tringle_indices] = True
```

```python
# Heatmap
plt.figure(figsize=[15,10])
sns.set_style('white')
sns.heatmap(corr,mask = mask, annot= False, annot_kws={'size': 14},
            cmap = 'PiYG', vmax=0.3, center=0, linewidths=0.5)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.show()
```
